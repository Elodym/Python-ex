# Pandas visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_upper = pd.DataFrame([[1,3], [0,2]], columns=['L','C'])
df_lower = pd.DataFrame([[-1,0],[5,4]], columns=['C','R'])
df_upper
```

```python
df_lower
```

```python
#Q1 stack df_lower below df_upper (method 1)
pd.concat([df_upper, df_lower], sort=False)
```

```python
#Q2 stack df_lower below df_upper (method 2)
df_upper.append(df_lower, sort=False)
```

```python
#different column ordering
df_upper.append(df_lower, sort=True)
```

```python
#Q3 stack df_lower to the right of df_upper
pd.concat([df_upper, df_lower], axis=1)
```

```python
#Q4 stack df_lower below df_upper with inner join
pd.concat([df_upper, df_lower], sort=False, join='inner')
```

```python
#outer join = default (not necessary to specify)
pd.concat([df_upper, df_lower], sort=False, join='outer')
```

```python
#Q5 combine dataframes by columns (i.e. stack df_lower to the right of df_upper) + inner join
pd.concat([df_upper, df_lower], axis=1, join='inner')
#no difference with ot witout inner attribute as no NaN value when stack by columns
```

```python
df_left = pd.DataFrame([['London', 'Yes'], ['Paris', 'Yes'], ['Frankfurt', 'No']], columns = ['City', 'Capital?'])
df_right = pd.DataFrame([['London', 'UK'], ['Paris', 'France'], ['Madrid', 'Spain']], columns = ['City', 'Country'])
df_left
```

```python
df_right
```

```python
#Q6 merge 2 df on 'City' value
pd.merge(df_left, df_right, on='City', how='outer')
```

```python
#Q7 left join
pd.merge(df_left, df_right, on='City', how='left')
```

```python
#Q8 right join
pd.merge(df_left, df_right, on='City', how='right')
```

```python
#Q9 full outer join
pd.merge(df_left, df_right, how='outer', sort=False)
#False: displays df_left and adds df_right to the right and bottom (order witin df_left unchanged)
```

```python
pd.merge(df_left, df_right, how='outer', sort=True)
#True: displays common columns first (City), then remaining columns in order of appearance in df_left, then df_right
#Looks like the rows are ordered alphabetically
```
