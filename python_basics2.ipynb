Q1. Modify the function sum(l,u) so that it computes and returns the sum of all integers between l and u as above only if l<u. If l>=u, the function should print 'First number should be smaller than the second!' and return None.

```python
def sum2(l,u):
    sum2=0
    if l<u:
        for i in range(l, u+1):
            sum2+=i
        return(sum2)
    else:
        print('First number should be smaller than the second!')
        return
```

```python
sum2(10, -12)
```

```python
sum(3, 0)
```

```python
print(sum2(3, -4))
```

```python
#same:
x = sum2(3, -4)
print(x)
```

## Variable scope

```python
#Q2
#variable scope refers to the places where you can see or access a variable
#global scope: any Python function or class defined in this notebook can access the variable
#local scope: variables defined inside a function or class are not global, only the function or class can see the variable

def func(a,b):
    a=2 #local location
    b=1 #local
    return a-b
print(func(a,b))
```

```python
a=6 #global location
b=2 #global
print(func(a,b)) #will always out a because locally a=2 and b=1, no matter of global value
print(a)
print(b)
```

Q3. Write a Python function that prints the first n Fibonacci numbers as a list. Recall that Fibonacci numbers form the following sequence: 1, 1, 2, 3, 5, 8,..., where the sequence starts with 1 and 1, and then each successive number is the sum of the two preceding ones. Your function should take n as the argument. E.g. if n=4, the function prints [1,1,2,3]. Test your function for n=5 and n=25.

```python
def fib(n):
    l=[1,1]
    if n==1:
        print([1])
    elif n==2:
        print(1)
    else:
        for i in range(2,n):
            last=l[-1]+l[-2]
            l.append(last)
        print(l)
```

```python
fib(5)
```

```python
fib(25)
```

Q4. Write a Python function that returns a list of the unique elements of a list given as the argument. So if the function receives the list [1, 'a', 'a', 0, 3,3,3], it returns [1,'a',0,3].

```python
def unique(l):
    u=[] #u is an empty list
    u.append(l[0]) #append the character is the first position of l to the end of u
    for i in range(1,len(l)): #for i in the range between 1 and the length of l
        if l[i] not in u: #if the character with position number i in l does not belong to u
            u.append(l[i]) #then append this character to u
    return u

#so u is an empty list
#append the first character in list l to u
#now for each successive position in list l, if a character (in position i) does not yet belong to u, append it to u
#this ensures that each character in only appended once to u
#so intuitively, we don't remove characters that appear multiple times in l
#but we successively add add unique characters to u
#i.e. create new unique list, don't change list l
```

```python
#this was just to understand that .append(l[0]) adds the character in first position in l
#forgot that [] in list meant position, nothing to do with the value 0
l=[2,3,7,'a']
u=[]
u.append(l[0])
print(u)
```

```python
#check:
print(unique([1,'a','a',0,3,3,3]))
```

Q5. Write a function that asks the user to input a string (a word) and then prints whether this string is a palindrome or not (a palindrome is a string that reads the same forward and backwards). Test this for the following words: toot, toast, deified, madam, and minimum.

```python
def pal(x):
    x=input('Enter a word:')
    y=x[::-1] #method 1 (simplified slicing code)
    if x==y:
        print('This word is a palindrome.')
    else:
        print('This string is not a palindrome.')
```

[::-1] interpretation: from beginning to end of list, take steps of -1 (backward) and print each value

[::2] interpretation: from beginning to end of list, take foward steps in 2. E.g. 'Elodie'. Output = 'Eoi'

```python
s = 'Elodie'
s[::2]
```
